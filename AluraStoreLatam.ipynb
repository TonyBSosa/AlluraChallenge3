### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

ingresos
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(ingresos.keys(), ingresos.values())
plt.title("Ingresos totales por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingresos")
plt.show()
```

# 2. Ventas por categoría

```python
categoria_tienda1 = tienda["Categoría del Producto"].value_counts()
categoria_tienda2 = tienda2["Categoría del Producto"].value_counts()
categoria_tienda3 = tienda3["Categoría del Producto"].value_counts()
categoria_tienda4 = tienda4["Categoría del Producto"].value_counts()

categoria_tienda1
```

```python
plt.figure()
categoria_tienda1.plot(kind="bar")
plt.title("Ventas por categoría - Tienda 1")
plt.xlabel("Categoría")
plt.ylabel("Cantidad vendida")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificaciones = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

calificaciones
```

```python
plt.figure()
plt.bar(calificaciones.keys(), calificaciones.values())
plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio")
plt.show()
```

# 4. Productos más y menos vendidos

```python
producto_mas_vendido = tienda["Producto"].value_counts().idxmax()
producto_menos_vendido = tienda["Producto"].value_counts().idxmin()

print("Producto más vendido:", producto_mas_vendido)
print("Producto menos vendido:", producto_menos_vendido)
```

```python
top_productos = tienda["Producto"].value_counts().head(10)

plt.figure()
top_productos.plot(kind="bar")
plt.title("Productos más vendidos - Tienda 1")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_promedio = {
    "Tienda 1": tienda["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}

envio_promedio
```

```python
plt.figure()
plt.bar(envio_promedio.keys(), envio_promedio.values())
plt.title("Costo promedio de envío por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo promedio de envío")
plt.show()
```


# Informe final
Después de analizar los datos de las cuatro tiendas, se pueden observar algunas diferencias importantes entre ellas. La Tienda 1 es la que genera más ingresos en total, lo que indica que tiene un mayor volumen de ventas en comparación con las demás. Además, las categorías más vendidas en esta tienda son muebles y electrónicos, lo que muestra que hay una fuerte demanda en ese tipo de productos. En cuanto a la satisfacción de los clientes, las calificaciones entre las tiendas son bastante similares, aunque la Tienda 3 tiene un promedio ligeramente más alto.

Por otro lado, la Tienda 4 es la que presenta los ingresos más bajos. Aunque tiene el costo de envío promedio más bajo, esto no parece ser suficiente para compensar la diferencia en ventas con respecto a las otras tiendas. Considerando todos los factores analizados, como los ingresos totales, las categorías más vendidas, las calificaciones de los clientes y el costo de envío, **la mejor decisión sería cerrar la Tienda 4, ya que es la que muestra el desempeño más bajo en comparación con las demás.**
